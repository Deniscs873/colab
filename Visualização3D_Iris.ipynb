```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from google.colab import files
import warnings # na versão atual existem vários warnins. vamos ignorar
warnings.filterwarnings("ignore")

import io
uploaded = files.upload() # uploaded é um dicionário, sendo a chave o nome do arquivo

# Loading dataset
# https://www.kaggle.com/datasets/uciml/iris - Basta fazer p download do arquivo CSV e carregar nesse código.
#list(uploaded.keys())[0] foi a forma que encontrei de não entrar com o nome do arquivo no código
iris = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

iris.head()
```

```python
iris.columns
```

```python
df=iris
```

```python
plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
fg = ax.scatter3D(df['SepalWidthCm'], df['SepalLengthCm'], df['PetalLengthCm'],s=df['PetalWidthCm']*50,c=df['PetalWidthCm'])

ax.set_xlabel('SepalWidthCm')
ax.set_ylabel('SepalLengthCm')
ax.set_zlabel('PetalLengthCm')

plt.colorbar(fg)
```

```python
import plotly.express as px
uploaded = files.upload() # uploaded é um dicionário, sendo a chave o nome do arquivo

# Loading dataset
# https://www.kaggle.com/datasets/uciml/Breast_cancer_data.csv - Basta fazer p download do arquivo CSV e carregar nesse código.
#list(uploaded.keys())[0] foi a forma que encontrei de não entrar com o nome do arquivo no código
df = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

df.head()
```

```python
figure = px.scatter_3d(df, x='mean_radius', y='mean_texture', z='mean_perimeter')
figure.show()
```

```python
import plotly.graph_objects as go
df.shape
figure = px.scatter_3d(df, x='mean_area',y = 'mean_radius',z= 'mean_perimeter')

figure.show()
```

```python
fig = go.Figure(data = [go.Surface(z=df.values)])
fig.update_layout(title= 'my graph',autosize=False, width=500, height= 500, margin = dict(l=65, r=50, b=65, t=98))

fig.show()
```
